```python
import pandas as pd

# Your actual DataFrame
df = pd.DataFrame({
    'Results_1_Metric': ['R2 LN', 'R2 LN'],
    'Results_1_Value': ['0.5618881510394962', '0.7212345678901234'],
    'Results_2_Metric': ['R2 LN', 'R2 LN'],
    'Results_2_Value': ['0.6312345678901234', '0.8212345678901234']
})

# Get odd columns containing numeric values
numeric_columns = [col for col in df.columns[::2] if df[col].dtype == 'object']

if numeric_columns:
    # Convert the numeric columns to numeric data type
    df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric)

    # Find the maximum value and its location
    max_value = df[numeric_columns].max().max()
    max_row, max_column = (df[numeric_columns] == max_value).stack().idxmax()

    # Display the results
    print(f'Highest Value: {max_value}')
    print(f'Row: {max_row}')
    print(f'Column: {max_column}')
else:
    print('No numeric columns found.')
```
